```python
class power_factor():
    
    
    def categorization(self,df):
        bins = [0, 0.75, 0.9, np.inf]
        names = ['Critical', 'Warning', 'Good']
        d = dict(enumerate(names, 1))
        df['Status'] = np.vectorize(d.get)(np.digitize(df['PF'], bins))
        return df
    
        
    def minimum(self,df):
        df=self.categorization(df)
        x=df.groupby(['EM_id'])['PF'].agg({'Minimum':'min','Maximum':'max','Average':'mean'})
        x.reset_index(inplace=True)
        return x
        
```

```python
cat = power_factor()
t=cat.minimum(df)
t
```
